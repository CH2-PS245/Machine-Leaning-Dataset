### CONTENT BASED

```python
import numpy as np
import pandas as pd
import random

dot_product = np.genfromtxt('/content/DishDash/ML/food-data/dot_prod.csv').astype(int)
preprocess_data = pd.read_csv('/content/DishDash/ML/food-data/preprocess_content_based_data.csv', sep=',')
food_img = food_img
```

```python
preprocess_data = preprocess_data.merge(food_img, on='food_id', how='left')
```

```python
preprocess_data
```

```python
preprocess_data.to_csv('preprocess_content_based_data.csv')
```

```python
def find_similar_food(food_ids):
    '''
    INPUT
    food_ids: List of Food IDs
    OUTPUT
    List of lists containing similar foods for each input food ID
    '''
    recommend = []

    for food_id in food_ids:
        food_idx = np.where(preprocess_data['food_id'] == food_id)[0][0]
        similar_idxs = np.where(dot_product[food_idx] == np.max(dot_product[food_idx]))[0]
        similar_food = list(zip(preprocess_data.iloc[similar_idxs, ]['food_id'], preprocess_data.iloc[similar_idxs, ]['food_name']))
        recommend.extend(similar_food)

    return recommend
```

```python
def get_image_path(food_id):
    return preprocess_data.loc[preprocess_data['food_id'] == food_id].reset_index(drop=True)['image'].iloc[0]
```

```python
food_ids = [3, 10, 2]

selected_food = [(id, preprocess_data.at[id-1, 'food_name'], get_image_path(id)) for id in food_ids]

similar_food_with_id = find_similar_food(food_ids)

# Extract the recommended food IDs and names
recommended_food_ids, recommended_food_names = zip(*similar_food_with_id)

flattened_similar_food = list(zip(recommended_food_ids, recommended_food_names))
random.shuffle(flattened_similar_food)

simplified = flattened_similar_food[:10]


for id, name, image_path in selected_food:
    print(f'Makanan Pilihan: {name}')
    print(f'Food ID: {id}')
    print(f'Image Path: {image_path}')

output_food_id = []
output_food_name = []
output_food_img = []
print('\n')
for id, name in simplified:
    image_path = get_image_path(int(id))
    output_food_id.append(id)
    output_food_name.append(name)
    output_food_img.append(image_path)
    print(f'Rekomendasi: {name}')
    print(f'Food ID: {int(id)}')
    print(f'Image Path: {image_path}')
```
